```python
def search_mercari_items(keyword, exclude_keywords):
    try:
        # Perform the search using function arguments and fixed parameters
        search_results = search(
            keyword,
            sort=MercariSort.SORT_SCORE,          # Fixed sort as per original code
            order=MercariOrder.ORDER_DESC,         # Fixed order as per original code
            status=MercariSearchStatus.ON_SALE,    # Fixed status as per original code
            exclude_keywords=exclude_keywords     # Use provided exclusion keywords
        )

        # Iterate through results and print
        for item in search_results:
            print("{}; {}; {}".format(item.productName, item.productURL, item.price))

    except Exception as e:
        print(f"An error occurred during the Mercari search: {e}")
        
search_mercari_items(keyword="iPhone15 Pro 256GB", exclude_keywords="max")
```

```python
def get_mercari_items(keyword, exclude_keywords, min_price=None, max_price=None):
    try:
        search_results = search(
            keyword,
            sort=MercariSort.SORT_PRICE,
            order=MercariOrder.ORDER_ASC,
            status=MercariSearchStatus.ON_SALE,
            exclude_keywords=exclude_keywords
        )

        items_printed = 0
        for item in search_results:
            try:
                price = float(item.price)
                min_check_passed = (min_price is None) or (price >= float(min_price))
                max_check_passed = (max_price is None) or (price <= float(max_price))

                if min_check_passed and max_check_passed:
                    print("{}; {}; {}".format(item.productName, item.productURL, item.price))
                    items_printed += 1
                elif max_price is not None and price > float(max_price):
                    break # Exit loop early since results are price-sorted

            except (ValueError, TypeError) as price_err:
                continue # Skip item with invalid price

        if items_printed == 0:
            print("No items found matching the search criteria and specified price range.")

    except Exception as e:
        print(f"An error occurred: {e}")

get_mercari_items(keyword="iPhone15 Pro 256GB", exclude_keywords="11 12 13 14 16 ジャンク max", min_price=100000, max_price=150000)
```
